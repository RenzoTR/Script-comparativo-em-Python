```python
#Pegando a biblioteca (pytrends)

from pytrends.request import TrendReq
# adicionar parametros
trend = TrendReq(hl='pt-BR', tz=180)
# Lista de palavras-chave
kw_list = ['Vivo', 'Claro']
```

```python
#Criando a aba de resposta
#trend.build_payload(kw_list)
trend.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='BR-ES', gprop='')
```

```python
#Dataframe do google trends
data = trend.interest_over_time()
```

```python
# importar o plotly
import plotly.express as px
```

```python
	px.line(data, x= data.index, y=['Vivo', 'Claro'], title = 'Vivo vs Claro at ES-Brazil')
```
